```python
#Function to convert genres and keywords dict to a list
def convert(obj):
    L = []
    
    for i in ast.literal_eval(obj):
        L.append(i['name'])
        
    return L
```

```python
# Applying the convert function
df_movies['genres'] = df_movies['genres'].apply(convert)
df_movies['keywords'] = df_movies['keywords'].apply(convert)

# Display five for beverity
df_movies.head()
```

```python
# Function to convert the 'cast' dict to a list
def convert_castt(obj):
    L = []
    counter = 0
    
    for i in ast.literal_eval(obj):
        if i != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
            
    return L
```

```python
df_movies['cast'] = df_movies['cast'].apply(convert)
df_movies.head()
```

```python
def convert_dict(obj):
    L = []
    if isinstance(obj, list):
        # If 'obj' is already a list, assume it's in the correct format
        return obj
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
df_movies['crew'] = df_movies['crew'].apply(convert_dict)
df_movies.head()
```

```python
# Now applying a lambda function to convert the 'overview' string to a compatible form.
df_movies['overview'] = df_movies['overview'].apply(lambda x: x.split() if type(x) is str else 'empty')
df_movies.head()
```

```python
df_movies['genres'] = df_movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
df_movies['keywords'] = df_movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
df_movies['cast'] = df_movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
df_movies['crew'] = df_movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
df_movies.head()
```

## Feature Engg. to  create a new column by using the existing ones

```python
df_movies['tags'] = (
    df_movies['overview'].astype(str) +
    df_movies['genres'].astype(str) +
    df_movies['keywords'].astype(str) +
    df_movies['cast'].astype(str) +
    df_movies['crew'].astype(str)
)
df_movies.head()
```

## Creating a seprate dataframe for more consises process
